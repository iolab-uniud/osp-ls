```python
import pandas as pd

# figures library
import matplotlib
matplotlib.use('agg')

import matplotlib.pyplot as plt
plt.rcParams.update({'pdf.fonttype': 42})

import seaborn as sns
sns.set(style='whitegrid',context="poster")
```

```python
def gap(original, our):
    return 100*(our-original)/original

def label(gap): 
    if gap > 0:
        return "damodaran"
    if gap < 0:
        return "our"
    return "equal"

def label_nuances(gap): 
    if gap > 1:
        return ">1"
    if gap < -1:
        return "< -1"
    if gap > 0:
        return " 0 to 1"
    if gap < 0:
        return "-1 to 0"
    return "equal"

def size(jobs):
    if jobs <= 20:
        return "small"
    if jobs <=100:
        return "medium"
    return "large"
```

```python
original = pd.read_csv("../processed-datasets/damodaran/damodaran-original.csv",index_col=0)
our = pd.read_csv("../processed-datasets/damodaran/damodaran-our-sa-time-wise.csv",index_col=0)
desc = pd.read_csv("../processed-datasets/damodaran/damodaran-instance-properties.csv")
df = pd.merge(original,our)
df = pd.merge(df,desc)
df["gap"] = df.apply(lambda row: gap(row["damodaran_mean"],row["our_sa_mean"]),axis=1)
df["label"] = df.apply(lambda row: label(row["gap"]),axis=1)
df["label_nuances"] = df.apply(lambda row: label_nuances(row["gap"]),axis=1)
df["size"] = df.apply(lambda row: size(row["Jobs"]),axis=1)
```
